```python
import matplotlib.pyplot as plt
import numpy as np
from dataset import LINK_LENGTHS

data = np.load('inference_results.npz')
theta_init = data['theta_init']
theta_final = data['theta_final']
target_pos = data['target_pos']
```

```python

def plot_arm(theta, color='b'):
    x, y, angle = 0, 0, 0
    xs, ys = [0], [0]
    for i in range(3):
        angle += theta[i]
        x += LINK_LENGTHS[i] * np.cos(angle)
        y += LINK_LENGTHS[i] * np.sin(angle)
        xs.append(x)
        ys.append(y)
    plt.plot(xs, ys, marker='o', color=color)

# Plote bra√ßo antes e depois
plot_arm(theta_init, color='gray')
plot_arm(theta_final, color='green')
plt.plot(*target_pos, 'rx', markersize=10, label='Alvo')
plt.axis('equal')
plt.grid()
plt.legend()
plt.show()
```
